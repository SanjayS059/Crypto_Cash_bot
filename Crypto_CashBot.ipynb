```python
import requests
import telebot

# Telegram bot setup
BOT_TOKEN = '7049240324:AAE6v5V0rwgEaR4NqE0ycup1_kA_YV6vfDo'
CHAT_ID = '924026889'
bot = telebot.TeleBot(BOT_TOKEN)

# NewsData.io API setup
API_KEY = 'pub_508623bb023f722e88fe1ec8891703cbf813b'
BASE_URL = 'https://newsdata.io/api/1/news'

def fetch_crypto_news():
    params = {
        'apikey': API_KEY,
        'q': 'crypto OR bitcoin OR ethereum OR blockchain',
        'language': 'en',
        'category': 'business,technology'
    }

    try:
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status()
        news_data = response.json()

        news_items = []
        for article in news_data.get('results', []):
            title = article.get('title')
            link = article.get('link')
            if title and link:
                news_items.append(f"{title}\n{link}\n\n")

        return news_items

    except requests.RequestException as e:
        print(f"An error occurred while fetching news: {e}")
        return []

def send_telegram_message(message):
    bot.send_message(CHAT_ID, message)

def main():
    news = fetch_crypto_news()
    if news:
        message = "Latest Crypto News:\n\n" + ''.join(news)
        send_telegram_message(message)
    else:
        send_telegram_message("No new crypto news found at this time.")

if __name__ == "__main__":
    main()
```
